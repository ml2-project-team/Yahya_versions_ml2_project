##Feature Engineering

```python
#feature engineering and solve the skewness both in one
#first one
# Apply log transformation to the Amount column
credit_df['Amount_log'] = np.log(credit_df['Amount'] + 1)


# Plot original and transformed distributions
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title('Original Amount Distribution')

plt.subplot(1, 2, 2)
sns.histplot(credit_df['Amount_log'], bins=50, kde=True, color='red')
plt.title('Log-Transformed Amount Distribution')

plt.tight_layout()
plt.show()

#Here is we will add a new column ("Amount_log") to the credit_df and secondly make a histogram plots for the original Amount and the Amount_log to show how the skewness is reduced. .
```

```python
#the second one of feature engineering is to reflect the hour when each transaction occurred....
#0 means that it is in the midnight 12

credit_df['Hour_of_transaction_in_day'] = (credit_df['Time'] // 3600) % 24

# We engineered the Hour_of_Day feature to capture the time of day each transaction occurred, assuming Time = 0 corresponds to midnight. This temporal feature enables unsupervised models to learn daily transaction patterns. For example, clustering algorithms can separate transactions during peak hours from those at unusual times, and anomaly detection models may flag late-night high-value transactions as outliers. Since fraud often happens outside typical working hours, this feature contributes to detecting anomalies without requiring access to labels.


#Now to have the benifit from this feature engineering is make another dataframe that contain the instance that the fraud occurs and to show in which hour the most fraud transactions happens......
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(x='Hour_of_transaction_in_day', data=credit_df, palette='viridis')

plt.title('Transaction Count by Hour of Day', fontsize=14)
plt.xlabel('Hour of Day (0 = Midnight)', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(ticks=range(24))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
