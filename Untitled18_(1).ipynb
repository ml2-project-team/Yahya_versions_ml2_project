```python
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA

# Separate features and target
X = credB.drop(columns=['Class'])  # or adjust if your target has a different name
y = credB['Class']  # save labels for later visualization

# Scale features using RobustScaler
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fit PCA to all components
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100, marker='o')
plt.axhline(y=90, color='r', linestyle='--', label='90% Variance')
plt.axhline(y=95, color='g', linestyle='--', label='95% Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Explained Variance by PCA Components')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
